```python
#Question:4
dic={
    "Name":"Kanwal Abid",
    "Father Name":"Muhammad Abid",
    "Age":28,
    "Profession":"Lecturer",
    "Height":5.4,
     "teaching Experience":12
}
sum=0
for i in dic.values():
    if type(i) == int or type(i) == float:
        sum=sum+i
print("Sum of numeric value:", sum)
    
```

```python
#Question:5
lst=input("Enter a list values:").split()
duplicate=[]
for i in lst:
    if lst.count(i) > 1 and i  not in duplicate:
        duplicate.append(i)
print("Duplicate values:", duplicate)
```
